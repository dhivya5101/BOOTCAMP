```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.signal import convolve2d
```

```python
# Load an image in grayscale
image = cv2.imread('dog.10.jpg', cv2.IMREAD_GRAYSCALE)

# Resize the image for consistent processing (optional)
image = cv2.resize(image, (128, 128))

# Normalize the image (optional but common in CNNs)
image = image / 255.0

# Display original image
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Define a simple edge-detection filter
filter = np.array([[-1, -1, -1], 
                   [-1,  8, -1], 
                   [-1, -1, -1]])

print("Filter Kernel:\n", filter)
```

```python
# Perform convolution using scipy's convolve2d function
feature_map = convolve2d(image, filter, mode='valid')

# Display feature map
plt.imshow(feature_map, cmap='gray')
plt.title("Feature Map after Convolution")
plt.axis('off')
plt.show()
```
