```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = X**3 + X**2 - X + np.random.normal(0, 3, size=X.shape)  # Cubic relationship with noise

lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_lin_pred = lin_reg.predict(X)

poly_features = PolynomialFeatures(degree=3)
X_poly = poly_features.fit_transform(X)

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_poly_pred = poly_reg.predict(X_poly)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='yellow', label="Data")
plt.plot(X, y_lin_pred, color='red', label="Linear Regression")
plt.plot(X, y_poly_pred, color='green', label="Polynomial Regression (Degree=3)")
plt.legend()
plt.title("Linear vs Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
